## 11. Data Cleaning and Preparation
- converting date columns to datetime
- converting numeric columns to numeric types
- handling invalid values
- creating date consistency flags
- creating geographic data quality flags
- summarizing before/after row counts and data types
- saving cleaned analysis-ready CSV files

```python
# Make working copies so the earlier cleaned dataframes stay unchanged
listed_wk45 = df_listed_clean.copy()
sold_wk45 = df_sold_clean.copy()

listed_original_rows = len(listed_wk45)
sold_original_rows = len(sold_wk45)

print("Starting listed rows:", listed_original_rows)
print("Starting sold rows:", sold_original_rows)
```

```python
def convert_existing_columns_to_datetime(df, columns):
    for col in columns:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")
    return df

def convert_existing_columns_to_numeric(df, columns):
    for col in columns:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")
    return df

date_cols = [
    "CloseDate",
    "PurchaseContractDate",
    "ListingContractDate",
    "ContractStatusChangeDate"
]

numeric_cols = [
    "ClosePrice",
    "ListPrice",
    "OriginalListPrice",
    "LivingArea",
    "LotSizeSquareFeet",
    "DaysOnMarket",
    "BedroomsTotal",
    "BathroomsTotalInteger",
    "Latitude",
    "Longitude"
]

listed_wk45 = convert_existing_columns_to_datetime(listed_wk45, date_cols)
sold_wk45 = convert_existing_columns_to_datetime(sold_wk45, date_cols)

listed_wk45 = convert_existing_columns_to_numeric(listed_wk45, numeric_cols)
sold_wk45 = convert_existing_columns_to_numeric(sold_wk45, numeric_cols)

print("Date and numeric conversions completed.")
```

```python
# Remove or filter clearly invalid numeric values required by the handbook.
# We only apply a rule if that column exists in the dataframe.

def apply_invalid_value_filters(df):
    mask = pd.Series(True, index=df.index)

    if "ClosePrice" in df.columns:
        mask &= df["ClosePrice"] > 0
    if "LivingArea" in df.columns:
        mask &= df["LivingArea"] > 0
    if "DaysOnMarket" in df.columns:
        mask &= df["DaysOnMarket"] >= 0
    if "BedroomsTotal" in df.columns:
        mask &= df["BedroomsTotal"].isna() | (df["BedroomsTotal"] >= 0)
    if "BathroomsTotalInteger" in df.columns:
        mask &= df["BathroomsTotalInteger"].isna() | (df["BathroomsTotalInteger"] >= 0)

    return df.loc[mask].copy()

listed_wk45 = apply_invalid_value_filters(listed_wk45)
sold_wk45 = apply_invalid_value_filters(sold_wk45)

print("Listed rows after invalid-value filtering:", len(listed_wk45))
print("Sold rows after invalid-value filtering:", len(sold_wk45))
```

```python
# Create date consistency flags.
# These help identify records where the transaction timeline looks logically incorrect.

def add_date_flags(df):
    if "ListingContractDate" in df.columns and "CloseDate" in df.columns:
        df["listing_after_close_flag"] = df["ListingContractDate"] > df["CloseDate"]
    else:
        df["listing_after_close_flag"] = False

    if "PurchaseContractDate" in df.columns and "CloseDate" in df.columns:
        df["purchase_after_close_flag"] = df["PurchaseContractDate"] > df["CloseDate"]
    else:
        df["purchase_after_close_flag"] = False

    if "ListingContractDate" in df.columns and "PurchaseContractDate" in df.columns:
        df["negative_timeline_flag"] = df["PurchaseContractDate"] < df["ListingContractDate"]
    else:
        df["negative_timeline_flag"] = False

    return df

listed_wk45 = add_date_flags(listed_wk45)
sold_wk45 = add_date_flags(sold_wk45)

listed_wk45[[
    "listing_after_close_flag",
    "purchase_after_close_flag",
    "negative_timeline_flag"
]].sum()
```

```python
# Create geographic data quality flags.

def add_geo_flags(df):
    if "Latitude" in df.columns and "Longitude" in df.columns:
        df["missing_coords_flag"] = df["Latitude"].isna() | df["Longitude"].isna()
        df["zero_coords_flag"] = (df["Latitude"] == 0) | (df["Longitude"] == 0)
        df["invalid_longitude_flag"] = df["Longitude"] > 0

        # Broad plausibility bounds for California-area coordinates
        df["out_of_bounds_flag"] = (
            (df["Latitude"] < 30) | (df["Latitude"] > 45) |
            (df["Longitude"] < -130) | (df["Longitude"] > -110)
        )
    else:
        df["missing_coords_flag"] = False
        df["zero_coords_flag"] = False
        df["invalid_longitude_flag"] = False
        df["out_of_bounds_flag"] = False

    return df

listed_wk45 = add_geo_flags(listed_wk45)
sold_wk45 = add_geo_flags(sold_wk45)

sold_wk45[[
    "missing_coords_flag",
    "zero_coords_flag",
    "invalid_longitude_flag",
    "out_of_bounds_flag"
]].sum()
```

```python
# Fill a few useful numeric fields with medians when they exist.

def fill_selected_numeric_nulls(df, columns):
    for col in columns:
        if col in df.columns and df[col].isnull().any():
            df[col] = df[col].fillna(df[col].median())
    return df

fill_cols = ["BedroomsTotal", "BathroomsTotalInteger", "DaysOnMarket"]

listed_wk45 = fill_selected_numeric_nulls(listed_wk45, fill_cols)
sold_wk45 = fill_selected_numeric_nulls(sold_wk45, fill_cols)

print("Selected missing numeric values filled with medians where applicable.")
```

```python
# Before/after row counts and type confirmations

print("LISTED before cleaning:", listed_original_rows)
print("LISTED after cleaning:", len(listed_wk45))
print("SOLD before cleaning:", sold_original_rows)
print("SOLD after cleaning:", len(sold_wk45))

print("\nListed dtypes:")
print(listed_wk45.dtypes)

print("\nSold dtypes:")
print(sold_wk45.dtypes)
```

```python
# Date consistency and geographic quality summaries

print("LISTED date flag counts")
print(listed_wk45[[
    "listing_after_close_flag",
    "purchase_after_close_flag",
    "negative_timeline_flag"
]].sum())

print("\nLISTED geographic flag counts")
print(listed_wk45[[
    "missing_coords_flag",
    "zero_coords_flag",
    "invalid_longitude_flag",
    "out_of_bounds_flag"
]].sum())

print("\nSOLD date flag counts")
print(sold_wk45[[
    "listing_after_close_flag",
    "purchase_after_close_flag",
    "negative_timeline_flag"
]].sum())

print("\nSOLD geographic flag counts")
print(sold_wk45[[
    "missing_coords_flag",
    "zero_coords_flag",
    "invalid_longitude_flag",
    "out_of_bounds_flag"
]].sum())
```

```python
# Save final cleaned datasets

listed_wk45.to_csv("cleaned_listed_week45.csv", index=False)
sold_wk45.to_csv("cleaned_sold_week45.csv", index=False)

print("Saved cleaned_listed_week45.csv")
print("Saved cleaned_sold_week45.csv")
```

## Cleaning + Validation

```python
# Save original row counts
original_sold_rows = len(df_sold_clean)
original_listed_rows = len(df_listed_clean)

print("Original sold rows:", original_sold_rows)
print("Original listed rows:", original_listed_rows)
```

```python
# 90% missing value threshold
# Checking if we dropped the columns correctly earlier
missing_pct_sold = df_sold_clean.isnull().mean()
missing_pct_listed = df_listed_clean.isnull().mean()

sold_drop_cols = missing_pct_sold[missing_pct_sold > 0.9].index
listed_drop_cols = missing_pct_listed[missing_pct_listed > 0.9].index

df_sold_clean = df_sold_clean.drop(columns=sold_drop_cols)
df_listed_clean = df_listed_clean.drop(columns=listed_drop_cols)

print("Dropped sold cols:", list(sold_drop_cols))
print("Dropped listed cols:", list(listed_drop_cols))
```

```python
# Convert date columns
date_cols = ['CloseDate','PurchaseContractDate','ListingContractDate','ContractStatusChangeDate']

for col in date_cols:
    if col in df_sold_clean.columns:
        df_sold_clean[col] = pd.to_datetime(df_sold_clean[col], errors='coerce')
    if col in df_listed_clean.columns:
        df_listed_clean[col] = pd.to_datetime(df_listed_clean[col], errors='coerce')
```

```python
# Numeric cleaning
numeric_cols = ['ClosePrice','LivingArea','DaysOnMarket','BedroomsTotal','BathroomsTotalInteger']

for col in numeric_cols:
    if col in df_sold_clean.columns:
        df_sold_clean[col] = pd.to_numeric(df_sold_clean[col], errors='coerce')
    if col in df_listed_clean.columns:
        df_listed_clean[col] = pd.to_numeric(df_listed_clean[col], errors='coerce')
```

```python
# Remove invalid values
df_sold_clean = df_sold_clean[
    (df_sold_clean['ClosePrice'] > 0) &
    (df_sold_clean['LivingArea'] > 0) &
    (df_sold_clean['DaysOnMarket'] >= 0)
]
```

```python
# Date consistency flags
df_sold_clean['listing_after_close_flag'] = df_sold_clean['ListingContractDate'] > df_sold_clean['CloseDate']
df_sold_clean['purchase_after_close_flag'] = df_sold_clean['PurchaseContractDate'] > df_sold_clean['CloseDate']
df_sold_clean['negative_timeline_flag'] = df_sold_clean['PurchaseContractDate'] < df_sold_clean['ListingContractDate']

print("Listing after close:", df_sold_clean['listing_after_close_flag'].sum())
print("Purchase after close:", df_sold_clean['purchase_after_close_flag'].sum())
print("Negative timeline:", df_sold_clean['negative_timeline_flag'].sum())
```

```python
# Geographic flags
df_sold_clean['missing_coords_flag'] = df_sold_clean['Latitude'].isnull() | df_sold_clean['Longitude'].isnull()
df_sold_clean['zero_coords_flag'] = (df_sold_clean['Latitude'] == 0) | (df_sold_clean['Longitude'] == 0)
df_sold_clean['invalid_longitude_flag'] = df_sold_clean['Longitude'] > 0

print("Missing coords:", df_sold_clean['missing_coords_flag'].sum())
print("Zero coords:", df_sold_clean['zero_coords_flag'].sum())
print("Invalid longitude:", df_sold_clean['invalid_longitude_flag'].sum())
```

```python
# Final row counts
print("Final sold rows:", len(df_sold_clean))
print("Final listed rows:", len(df_listed_clean))
```
